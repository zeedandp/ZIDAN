**Bukti bahwa** \( T(v_1, v_2) = (v_1 + v_2, v_1) \) **adalah transformasi linier**

**Syarat transformasi linier:**
\begin{align*}
(1) \quad & T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) \\
(2) \quad & T(c\mathbf{u}) = cT(\mathbf{u})
\end{align*}

**Misalkan:**  
$$
\mathbf{u} = (u_1, u_2), \quad \mathbf{v} = (v_1, v_2), \quad c \in \mathbb{R}
$$


---

**Periksa syarat (1):**

\begin{align*}
T(\mathbf{u} + \mathbf{v}) &= T((u_1 + v_1, u_2 + v_2)) \\
&= ((u_1 + v_1) + (u_2 + v_2), u_1 + v_1) \\
&= (u_1 + u_2 + v_1 + v_2, u_1 + v_1)
\end{align*}

\begin{align*}
T(\mathbf{u}) + T(\mathbf{v}) &= (u_1 + u_2, u_1) + (v_1 + v_2, v_1) \\
&= (u_1 + u_2 + v_1 + v_2, u_1 + v_1)
\end{align*}

$$
\Rightarrow T(\mathbf{u} + \mathbf{v}) = T(\mathbf{u}) + T(\mathbf{v}) \quad \text{(✓)}
$$

---

**Periksa syarat (2):**

\begin{align*}
T(c\mathbf{u}) &= T((cu_1, cu_2)) \\
&= (cu_1 + cu_2, cu_1) \\
&= c(u_1 + u_2, u_1) \\
&= cT(\mathbf{u}) \quad \text{(✓)}
\end{align*}

---

**Kesimpulan:**  
\( T \) adalah transformasi linier.
