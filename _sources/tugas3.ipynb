# Sistem Persamaan Linear dan Invers Matriks

Misalkan kita memiliki sistem persamaan:

\[
AX = B
\]

dengan:
$$
A =
\begin{bmatrix}
2 & 1 & -1 \\
1 & 3 & 2 \\
1 & -1 & 2
\end{bmatrix}
$$

$$
B =
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$

Invers dari \( A \) adalah:

$$
A^{-1} =
\begin{bmatrix}
\frac{7}{7} & \frac{-1}{7} & \frac{-5}{7} \\
\frac{-1}{7} & \frac{3}{7} & \frac{2}{7} \\
\frac{-5}{7} & \frac{2}{7} & \frac{6}{7}
\end{bmatrix}
$$

$$
X = A^{-1} B =
\begin{bmatrix}
1 \\
0 \\
0
\end{bmatrix}
$$
